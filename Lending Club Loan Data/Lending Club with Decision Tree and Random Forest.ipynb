# Setting up the Data

```python
loans.info()
```

```python
loans.head()
```

## Convert Categorical Features

```python
categorical_features = ["purpose"]
```

```python
final_data = pd.get_dummies(data=loans, columns=categorical_features, drop_first=True)
```

```python
final_data.info()
```

```python
final_data.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop("not.fully.paid", axis=1)
y = final_data["not.fully.paid"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Predicting Model

## Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

## Predictions and Evaluation of Decision Tree
**Predict the class of not.fully.paid for the X_test data.**

```python
prediction_dtree = dtree.predict(X_test)
```

#### Confusion Matrix and Classification Report

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, prediction_dtree))
```

```python
confusion_matrix(y_test, prediction_dtree)
```

## Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X_train, y_train)
```

## Predictions and Evaluation of Random Forest

**Predict the class of not.fully.paid for the X_test data.**

```python
prediction_rfc = rfc.predict(X_test)
```

**Classification Report**

```python
print(classification_report(y_test, prediction_rfc))
```

**Confusion Matrix**

```python
confusion_matrix(y_test, prediction_rfc)
```

## Conclusion: 
**Random Forest is better than Decision Tree.**
